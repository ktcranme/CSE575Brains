```python
import numpy as np
import pandas as pd
import scipy.io as sio
import os
import pickle
```

```python
path = r'C:\Users\deves\Downloads\CSE575Brains-master\brainnetworks\smallgraphs'
path
```

```python
data_list = []
for r, d, f in os.walk(path):
    for file in f:
        mat_file = sio.loadmat(path + '\\' + file)
        data = mat_file['fibergraph'].todense()
        data = data.flatten()
        data = data.tolist()
        data[0].append(file[:-10])
        data_list.append(data[0])
```

```python
df = pd.DataFrame(data_list)
df.head(10)
```

```python
meta_data_path = r"C:\Users\deves\Downloads\CSE575Brains-master\brainnetworks\metainfo.xls"
meta_data_df = pd.read_excel(meta_data_path)
meta_data_df = meta_data_df.drop(columns = ['Unnamed: 12'])
meta_data_df.head(10)
```

```python
final_df = pd.merge(left=df, right=meta_data_df, left_on = 4900, right_on = 'URSI')
final_df.head(10)
```

```python
final_df.to_pickle('Brain_data.pkl')
```
